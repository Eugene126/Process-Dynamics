```python
import sympy as sp
sp.init_printing()
```

```python
D,h,H,d = sp.symbols('D,h,H,d', real = True )
```

```python
R = D/2
r = d/2
```

```python
Hp = sp.sqrt(H**2 - (R - r)**2)

radius = r + h/Hp*(R - r)
area = sp.pi * radius ** 2
volume = sp.integrate(area, (h,0,h))
volume
```

Getting the equation in terms of volume V(h) ---> h(V)

```python
V = sp.Symbol('V')
```

V = volume

```python
hV = sp.solve(V - volume, h)
```
