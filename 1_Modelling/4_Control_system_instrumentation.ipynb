# Instrumentation

A key aspect of operating a plant, let alone implementing control, is the ability to measure physical parameters. In this section we will discuss ways in which we can take measurements, relay the measurements to a controller and then take action based on the measurement. The figure below shows the three key elements required to implement a control system in a process. Once the control system has been implemented the system becomes a **closed-loop**. This means that the output from the process, for instance a flow rate is fed to a controller and based on that value an input is relayed to an actuator to adjust a valve which changes the flow rate. Therefore the process as a whole (plant and control system) are essentially a closed loop since the output effects the input and input affects the output. **Open-loop** refers to a system where the output does not effect the input.  

<img src = 'connection.png' width = 300px/>

The below example is for a feedback control system. In order to adjust the heating duty of the element the temperature needs to be measured. This is done with a thermocouple which reacts to a temperature and outputs a millivolt electrical signal, this is then amplified to a voltage level and transmitted to the controller. The controller then calculates the adjustment required and this is then transmitted to the final control element that adjusts the rate of heat transfer to the tank.

<img src = 'heating_system.png' width = 450px/>

In most cases it is ideal to have **on-line** measurement. This means a measurement that is transmitting a signal repetitively every few seconds or minutes. The alternative would be taking a sample once a day or hour. This would often be the case for a safety check or when on-line instrumentation is not available or too expensive for the process. The analysis would then be done in a lab and relayed back to the plant.

### Signals

The transmission of signals between control elements are either digital or analogue. Although most new instrumentation is based on digital technology, analogue instrumentation is still widely used. Industry standard analogue signals are either pneumatic, current or voltage signals.

<img src = "avsd.png" width = 450px/>

|Signal type|Range|
|:-:|-|
| Pneumatic signal   | 3 - 15 psig|
|Current signal| 4 - 20 mA|
|Voltage signal| 1 - 5 V|

The range of the output correspond to a range of the physical value. It is ideal if the relationship between the measurement signal and the physical value is linear. The physical value has to be within the range of the sensor for it to be able to take a measurement. In most cases the relationship is direct acting, for instance an increase in signal value relates to an increase in temperature. The opposite will be used when it is safer for process operation.

|Input|Output|
|:-:|:-:|
|50 $^\circ C$| 4 mA|
|150 $^\circ C$| 20 mA|

<img src = 'signal_linear.png' />

This can be converted to a equation that relates input and output of the sensor as follows:

$$T_m(mA) = \frac{20 mA - 4 mA}{150^\circ C - 50 ^\circ C} (T - 50 ^\circ C) + 4 mA$$
$$= 0.16 \frac{mA}{^\circ C} T - 4 mA$$

The gain of this sensor would be $0.16 \frac{mA}{^\circ C}$ and this would be constant for a linear instrument, however this is not always the case, for a nonlinear instrument the operating point will have to be know to convert the signal to a physical reading.

- Performance
- Reliability
- Materials of construction
- Prior use (training, spares)
- Potential for releasing process material to the environment
- Electrical classification (enclosure)

## Final control elements

This is the device that enables the process variable to be manipulated. In most of our cases the final control element will be a control valve that will manipulate the flow of material. This material can be liquid, solid or gas as a product or a rate of energy transfer.

#### Control valves

There are many types of control valves and each have there specific application:
* Globe valves
* Ball valves
* Disc valves
* Butterfly valves

<img src = 'control_valve.png' width = 650px/>

The choice of a control valve with be based on: 
* Operating pressure
* Operating temperature
* Operating fluid
* Ability to seal off
* Flow characteristics 

Control valves can be operated in one of two ways, air to open (A/O) or air to close (A/C). The choice of function will be likely based on the use of the control valve and safety considerations.

The flow characteristic of a control valve is the relationship between the fractional level the seat has been lifted and the opening of the valve. The three most common are:
<img src= "characteristics.png"/>

- Linear: $f = l$
- Quick opening $f = \sqrt{l}$
- Equal percentage $f = R^{l - 1}$

The flow through a control valve is describe with this equation:

$$F = C_v f(l) \sqrt{\frac{\Delta P}{g_s}}$$

$F$ is the volumetric flow rate, $C_v$ is the control valve coefficient, $f(l)$ is the flow characteristic, $l$ is the fractional opening and $g_s$ is the specific gravity of the fluid. Therefore the flow rate through a control valve is a function of both the valve opening and the pressure drop. For this reason the sizing of a control valve is very important. The pressure drop may vary when the flow rate varies.

<img src = 'installed.png' width = 300px/>

If the $\frac{\Delta P_{Eq}}{\Delta P_v}$ remains fairly constant then you would want to use a linear flow characteristic, otherwise a equal percentage characteristic would be used to counter act the variation in $\frac{\Delta P_{Eq}}{\Delta P_v}$.

#### Valve rangeability

Is the fraction of the flow range over which effective control is possible. If a valve is said to be able to function between 0.05 < $f(l)$ < 0.95, then is has a rangeability of $\frac{0.95}{0.05} = 19$

### Valve actuators

This is the element that will implement the valve opening, it is often beneficial if there is a valve positioner install that will check the position of the valve position and compare it to the set point and then make an adjustment. This will help to get rid of hysteresis.

Different types of actuators include: 
- Spring and diaphragm
- Pneumatic piston
- Electrical motor
- Hydraulic or electro-hydraulic
